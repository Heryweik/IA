```python
#Ejercicio 1: según el dataset de adidas, para las columnas categóricas, encontrar: 
#aquellas columnas que tienen mas de un nivel, y que por lo menos uno de esos niveles represente mas o igual del 98%.
#Para las columnas que cumplan esa premisa, eliminaran del array de columnas validas.
```

```python
#No he quitado las columnas numericas dado que en el enunciado nos dice que trabajemos directamente con "adidas.columns"

cols_validas = []
cols_many_levels = []
cols_one_level = []

size_df = adidas.shape[0]
umbral = .98

for c_ in adidas.columns:
    
    niveles = adidas.loc[ : , [c_] ].drop_duplicates().shape[0]
    
    if ( niveles / size_df ) >= umbral:
        cols_many_levels.append(c_)
    elif niveles == 1:
        cols_one_level.append(c_)
    else:
       cols_validas.append(c_)
```

```python
#Ejercicio 2: Según el dataset de adidas, para las columnas categóricas, encontrar aquellas columnas que tengan
#entre 3-6 niveles. Al nivel con la frecuencia mas baja lo vamos a remplazar por la palabra Others, 
#una vez ya tenga esa regla aplicarla al dataset original, tal cual como se hizo con el ejercicio hecho en clases.
```

```python
cols_validas2 = []
cols_many_levels2 = []
cols_more_level = []

size_df = adidas.shape[0]
umbral = .98

for c_ in adidas.columns:
    
    niveles = adidas.loc[ : , [c_] ].drop_duplicates().shape[0]
    
    if ( niveles / size_df ) >= umbral:
        cols_many_levels2.append(c_)
    elif (niveles >= 3) & (niveles <= 6):
        cols_more_level.append(c_)
    else:
       cols_validas2.append(c_) 
```

```python
e2 = pd.DataFrame()
```

```python
e2= adidas.groupby(["category"], as_index = False).agg({ 'sku' : ['count']} )
e2.columns = e2.columns.droplevel(1)
```

```python
e2 = e2.sort_values(["category"], ascending = False)
```

```python
e2
```

```python
e2["Porcentaje"] = e2["sku"] / size_df
```

```python
adidas.loc[ :  , "category"].value_counts() / size_df #Accessories es el nivel con la frecuencia mas baja
```

```python
e2
```

```python
e2["category_FB"] = 'Others'
```

```python
e2["category_FB"] = e2.apply( lambda r: r["category"] if r["Porcentaje"] > 0.1 else r["category_FB"], axis = 1 )
```

```python
adidas = adidas.merge(e2, on=["category"], how="left") #Aplicamos al DataSet original
```
